```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
```

```python
# Function to load and process files based on the given conditions
def process_files(file_list, column_order, rename_dict, drop_column=None):
    dfs = []
    for file in file_list:
        df = pd.read_csv(file, delimiter=';', on_bad_lines='skip', dtype=str)
        if "Accident_Id" in df.columns:
            df.rename(columns={"Accident_Id": "Num_Acc"}, inplace=True)
        if drop_column and drop_column in df.columns:
            df.drop(columns=[drop_column], inplace=True)
        df = df.reindex(columns=column_order)
        dfs.append(df)
    combined_df = pd.concat(dfs, ignore_index=True)
    combined_df.rename(columns=rename_dict, inplace=True)
    return combined_df
```

```python
# Define file paths
source_folder = "source/"
caracteristiques_files = [os.path.join(source_folder, f) for f in os.listdir(source_folder) if f.startswith("caracteristiques")]
lieux_files = [os.path.join(source_folder, f) for f in os.listdir(source_folder) if f.startswith("lieux")]
usagers_files = [os.path.join(source_folder, f) for f in os.listdir(source_folder) if f.startswith("usagers")]
vehicules_files = [os.path.join(source_folder, f) for f in os.listdir(source_folder) if f.startswith("vehicules")]
```

```python
# Define column orders and rename dictionaries
caracteristiques_columns = ["Num_Acc", "jour", "mois", "an", "hrmn", "lum", "dep", "com", "agg", "int", "atm", "col", "adr", "lat", "long"]
caracteristiques_rename = {"Num_Acc": "AccID", "jour": "day", "mois": "month", "an": "year", "hrmn": "time", "lum": "lum", 
                           "dep": "dep_code", "com": "com_code", "agg": "location", "int": "int", "atm": "atm_condition", 
                           "col": "collision_type", "adr": "address", "lat": "lat", "long": "long"}

lieux_columns = ["Num_Acc", "catr", "voie", "v1", "v2", "circ", "nbv", "vosp", "prof", "pr", "pr1", "plan", "lartpc", 
                 "larrout", "surf", "infra", "situ", "vma"]
lieux_rename = {"Num_Acc": "AccID", "catr": "route_category", "voie": "route_number", "v1": "route_number_index1", 
                "v2": "alph_route_index", "circ": "traffic_regime", "nbv": "total_number_lanes", "vosp": "reserved_lane_code", 
                "prof": "longitudinal_profile", "pr": "upstream_terminal_number", "pr1": "distance_upstream_terminal", 
                "plan": "plan", "lartpc": "width_central_reservation", "larrout": "width_roadway", "surf": "surface_condition", 
                "infra": "infra", "situ": "accident_situation", "vma": "maximum_speed"}

usagers_columns = ["Num_Acc", "id_vehicule", "num_veh", "place", "catu", "grav", "sexe", "an_nais", "trajet", "secu1", 
                   "secu2", "secu3", "locp", "actp", "etatp"]
usagers_rename = {"Num_Acc": "AccID", "id_vehicule": "vehicleID", "num_veh": "num_veh", "place": "seat", "catu": "user_category", 
                  "grav": "gravity", "sexe": "gender", "an_nais": "birth_year", "trajet": "reason_travel", "secu1": "safety_equipment1", 
                  "secu2": "safety_equipment2", "secu3": "safety_equipment3", "locp": "pedestrian_location", "actp": "pedestrian_action", 
                  "etatp": "pedestrian_involved"}

vehicules_columns = ["Num_Acc", "id_vehicule", "num_veh", "senc", "catv", "obs", "obsm", "choc", "manv", "motor", "occutc"]
vehicules_rename = {"Num_Acc": "AccID", "id_vehicule": "vehicleID", "num_veh": "num_veh", "senc": "traffic_direction", 
                    "catv": "vehicle_category", "obs": "fixed_obstacle", "obsm": "mobile_obstacle", "choc": "initial_impact_point", 
                    "manv": "manv", "motor": "motor", "occutc": "number_occupants_publictransport"}
```

```python
# Process each group of files
characteristics = process_files(caracteristiques_files, caracteristiques_columns, caracteristiques_rename)
locations = process_files(lieux_files, lieux_columns, lieux_rename)
users = process_files(usagers_files, usagers_columns, usagers_rename, drop_column="id_usager")
vehicles = process_files(vehicules_files, vehicules_columns, vehicules_rename)
```

```python
# Convert 'lat' and 'long' fields to float in the 'characteristics' dataframe
characteristics['lat'] = characteristics['lat'].str.replace(',', '.').astype(float)
characteristics['long'] = characteristics['long'].str.replace(',', '.').astype(float)

# Convert 'time' field to proper time format in the 'characteristics' dataframe
characteristics['time'] = pd.to_datetime(characteristics['time'], format='%H:%M').dt.time

# Convert 'day', 'month', and 'year' fields to integer in the 'characteristics' dataframe
characteristics['day'] = pd.to_numeric(characteristics['day'], errors='coerce').astype(pd.Int32Dtype())
characteristics['month'] = pd.to_numeric(characteristics['month'], errors='coerce').astype(pd.Int32Dtype())
characteristics['year'] = pd.to_numeric(characteristics['year'], errors='coerce').astype(pd.Int32Dtype())
```

```python
# Replace invalid literals with NaN and convert specified columns in 'locations' dataframe
locations['total_number_lanes'] = pd.to_numeric(locations['total_number_lanes'].replace('#ERREUR', np.nan), errors='coerce').astype(pd.Int32Dtype())
locations['distance_upstream_terminal'] = pd.to_numeric(locations['distance_upstream_terminal'].str.replace(',', '.').replace('#ERREUR', np.nan), errors='coerce')
locations['width_central_reservation'] = pd.to_numeric(locations['width_central_reservation'].str.replace(',', '.').replace('#ERREUR', np.nan), errors='coerce')
locations['width_roadway'] = pd.to_numeric(locations['width_roadway'].str.replace(',', '.').replace('#ERREUR', np.nan), errors='coerce')
locations['maximum_speed'] = pd.to_numeric(locations['maximum_speed'].replace('#ERREUR', np.nan), errors='coerce').astype(pd.Int32Dtype())

#Drop fields with a large number of missing value
locations.drop(columns=['width_central_reservation'])
locations.drop(columns=['alph_route_index'])
```

```python
# Convert 'birth_year' field to integer and create 'age' column in 'users' dataframe
current_year = datetime.now().year
users['birth_year'] = pd.to_numeric(users['birth_year'].replace('#ERREUR', np.nan), errors='coerce').astype(pd.Int32Dtype())
users['age'] = current_year - users['birth_year']
```

```python
# Convert 'number_occupants_publictransport' field to integer in the 'vehicles' dataframe
vehicles['number_occupants_publictransport'] = pd.to_numeric(
    vehicles['number_occupants_publictransport'].replace('#ERREUR', np.nan), errors='coerce').astype(pd.Int32Dtype())

#Drop fields with a large number of missing value
vehicles.drop(columns=['number_occupants_publictransport'])
```

```python
# Function to add average line
def add_average_line(ax, data, color='red'):
    avg = data.mean()
    ax.axhline(avg, color=color, linestyle='--', linewidth=1)
    ax.text(len(data) - 1, avg, f'Avg: {avg:.2f}', color=color, ha='right')
```

```python
# Visualization 01: Distribution of Accidents by Year
plt.figure(figsize=(10, 6))
sns.countplot(data=characteristics, x='year', color='#66b3ff')
plt.title('Distribution of Accidents by Year')
plt.xlabel('Year')
plt.ylabel('Number of Accidents')
add_average_line(plt.gca(), characteristics['year'].value_counts())
plt.show()

# Analysis
print("The distribution of accidents by Year shows a significant drop in the number of accidents in 2020 (COVID-19 pandemic?), with a return to pre-pandemic levels in 2021 and 2022.")

# Validation
yearly_accidents = characteristics['year'].value_counts().sort_index()
print(yearly_accidents)
```

```python
# Visualization 02: Distribution of Accidents by Month
plt.figure(figsize=(10, 6))
sns.countplot(data=characteristics, x='month', color='#66b3ff')
plt.title('Distribution of Accidents by Month')
plt.xlabel('Month')
plt.ylabel('Number of Accidents')
add_average_line(plt.gca(), characteristics['month'].value_counts())
plt.show()

# Analysis
print("The distribution of accidents by month shows the seasonality in accidents, with higher frequencies potentially during summer vacation months and winter months due to weather conditions. But is not absolutely clear.")

# Validation
monthly_accidents = characteristics['month'].value_counts().sort_index()
print(monthly_accidents)
```

```python
# Visualization 03: Distribution of Accidents by Day
plt.figure(figsize=(10, 6))
day_counts = characteristics['day'].value_counts().sort_index()

# Create a DataFrame from the value counts for plotting
day_counts_df = pd.DataFrame({'Day': day_counts.index, 'Accidents': day_counts.values})

# Convert 'Day' to numeric, coercing errors to NaN and then dropping those rows
day_counts_df['Day'] = pd.to_numeric(day_counts_df['Day'], errors='coerce')
day_counts_df = day_counts_df.dropna(subset=['Day'])

# Ensure 'Day' is integer if it represents days
day_counts_df['Day'] = day_counts_df['Day'].astype(int)

sns.lineplot(data=day_counts_df, x='Day', y='Accidents', color='#66b3ff')
plt.title('Distribution of Accidents by Day')
plt.xlabel('Day')
plt.ylabel('Number of Accidents')
# Ensure add_average_line function is correctly defined and used
add_average_line(plt.gca(), day_counts)
plt.show()

# Analysis
print("The graph shows accidents are relatively stable, peaking around the 5th day and decreasing sharply after the 20th day. The drop on the 31st is justified by the fact that there are only 7 months with 31 days.")

# Validation
daily_accidents = characteristics['day'].value_counts().sort_index()
print(daily_accidents)
```

```python
# Visualization 04: Distribution of Accidents by Lighting Conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=characteristics, x='lum', color='#66b3ff', order=characteristics['lum'].value_counts().index)
plt.title('Distribution of Accidents by Lighting Conditions')
plt.xlabel('Lighting Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Analysis
print("The graph shows the distribution of accidents by lighting conditions. Most accidents occur under light condition 1 (Daylight), with nearly 145,000 incidents.")

# Validation
light_conditions_accidents = characteristics['lum'].value_counts().sort_values(ascending=False)
print(light_conditions_accidents)

```

```python
# Visualization 05: Distribution of Accidents by Weather Conditions
plt.figure(figsize=(10, 6))
sns.countplot(data=characteristics, x='atm_condition', color='#66b3ff', order=characteristics['atm_condition'].value_counts().index)
plt.title('Distribution of Accidents by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Number of Accidents')
plt.show()

# Analysis
print("The graph shows the distribution of accidents by weather conditions. Most accidents occur under weather condition 1 (Normal), with nearly 174,000 incidents.")

# Validation
weather_conditions_accidents = characteristics['atm_condition'].value_counts().sort_values(ascending=False)
print(weather_conditions_accidents)
```

```python
# Visualization 06: Function to group hours into 2-hour intervals
def time_to_interval(hour):
    return f"{int(hour//2*2):02d}:00 - {int(hour//2*2 + 2):02d}:00"

# Ensure 'time' field has correct format and extract hour
characteristics['time'] = pd.to_datetime(characteristics['time'], format='%H:%M:%S', errors='coerce')
characteristics['hour'] = characteristics['time'].dt.hour

# Apply the function to create 2-hour intervals
characteristics['time_interval'] = characteristics['hour'].apply(time_to_interval)

# Ensure the 'day' column is properly formatted
characteristics['day'] = pd.to_numeric(characteristics['day'], errors='coerce')

# Distribution of Accidents by Day and Time Interval (heatmap)
plt.figure(figsize=(14, 8))
accidents_day_time_interval = pd.crosstab(characteristics['day'], characteristics['time_interval'])
sns.heatmap(accidents_day_time_interval, cmap='coolwarm')  
plt.title('Distribution of Accidents by Day and Time Interval')
plt.xlabel('Time Interval')
plt.ylabel('Day')
plt.xticks(rotation=45)
plt.show()

# Analysis
print("Overall, the heatmap highlights that accidents are most frequent during the late afternoon to early evening hours and least frequent during the early morning hours, with a consistent distribution across the days of the month.")
```

```python
# Visualization 08: Top 10 regions by the number of accidents

# Group by region (department code) and count the number of accidents
accidents_by_region = characteristics.groupby('dep_code').size().reset_index(name='number_of_accidents')

# Sort by number of accidents in descending order and select the top 10
top_10_regions = accidents_by_region.sort_values(by='number_of_accidents', ascending=False).head(10)

# Plot the number of accidents by region
plt.figure(figsize=(14, 7))
plt.bar(top_10_regions['dep_code'], top_10_regions['number_of_accidents'], color='#66b3ff')
plt.title('Top 10 Regions by Number of Accidents')
plt.xlabel('Region (Department Code)')
plt.ylabel('Number of Accidents')
plt.tight_layout()
plt.show()

# Analysis
print("The graph shows the top 10 regions by the number of accidents, sorted in descending order.")

# Validation
top_10_regions
```

## Possible Biases in the Dataset:

- **Reporting Bias:** Not all accidents may be reported, leading to the underrepresentation of minor accidents or those involving certain demographics.
- **Selection Bias:** The dataset might over-represent accidents from certain regions, times of day, or types of roads.
- **Survivorship Bias:** Severe accidents with fatalities might be more thoroughly recorded, skewing the data towards more serious incidents.
- **Measurement Bias:** Inconsistent methods of data collection or recording errors can introduce bias, especially if different regions or departments use different standards.
- **Cultural and Societal Bias:** Cultural factors might influence the reporting and recording of accidents (e.g., reluctance to report minor incidents involving certain demographics).
