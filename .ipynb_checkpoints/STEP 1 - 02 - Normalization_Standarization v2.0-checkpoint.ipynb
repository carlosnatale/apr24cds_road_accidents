```python
# Import libraries needed to execute the code
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import offsetbox
warnings.filterwarnings('ignore')
import plotly.graph_objects as go
from sklearn.utils import resample
from matplotlib.image import imread
from sklearn.decomposition import PCA
from sklearn.datasets import make_swiss_roll
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.manifold import LocallyLinearEmbedding, Isomap, TSNE

%matplotlib inline
```

```python
# Import the clean data
data = pd.read_csv('source/data.csv', low_memory=False)
```

```python
# Copy of the original dataset for feature engineering and preprocessing
data_processed = data.copy()
```

```python
# 1. Feature Engineering: Create new time-based features
data_processed['hour'] = data_processed['time'] // 1000000
data_processed['day_of_week'] = pd.to_datetime(data_processed[['year', 'month', 'day']]).dt.dayofweek
```

```python
# 2. Encoding: Convert categorical variables to numerical formats
categorical_columns = data_processed.select_dtypes(include=['object']).columns

label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    data_processed[col] = le.fit_transform(data_processed[col])
    label_encoders[col] = le
```

```python
# 3. Normalization/Standardization: Normalize or standardize numerical features
numerical_columns = data_processed.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()
data_processed[numerical_columns] = scaler.fit_transform(data_processed[numerical_columns])
```

```python
# 4. Data Balancing: Check for class imbalance in the target variable
target_variable = 'gravity'
class_distribution = data_processed[target_variable].value_counts()

# Resample the dataset if class imbalance is detected 
if class_distribution.min() < 0.5 * class_distribution.max():
    minority_class = class_distribution.idxmin()
    majority_class = class_distribution.idxmax()
    
    df_majority = data_processed[data_processed[target_variable] == majority_class]
    df_minority = data_processed[data_processed[target_variable] == minority_class]
    
    df_minority_upsampled = resample(df_minority, 
                                     replace=True,     # sample with replacement
                                     n_samples=class_distribution.max(),    # to match majority class
                                     random_state=42)  # reproducible results
    
    data_balanced = pd.concat([df_majority, df_minority_upsampled])
else:
    data_balanced = data_processed
```
