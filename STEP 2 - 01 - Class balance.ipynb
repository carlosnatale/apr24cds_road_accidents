```python
import warnings
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, LabelEncoder
```

```python
data = pd.read_csv('source/data.csv')
```

```python
# List of columns to convert to object type
columns_to_convert = [
    'AccID', 'accident_situation', 'atm_condition', 'collision_type', 'com_code', 'dep_code', 
    'fixed_obstacle', 'gender', 'gravity', 'infra', 'initial_impact_point', 'int', 'location', 
    'longitudinal_profile', 'lum', 'manv', 'mobile_obstacle', 'motor', 'num_veh_x', 'num_veh_y',
    'pedestrian_action', 'pedestrian_location', 'plan', 'reason_travel', 'reserved_lane_code', 
    'route_category', 'safety_equipment1', 'safety_equipment2', 'safety_equipment3', 'seat', 
    'surface_condition', 'time', 'traffic_direction', 'traffic_regime', 'upstream_terminal_number', 
    'user_category', 'vehicle_category', 'vehicleID_x', 'vehicleID_y'
]
```

```python
# Convert specified columns to object type
data[columns_to_convert] = data[columns_to_convert].astype('object')
```

```python
# Convert 'time' field to proper time format in the 'data' dataframe
data['time'] = pd.to_datetime(data['time'], format='%H:%M:%S').dt.time
```

```python
# Find all fields with the value '-1'
fields_with_minus_one = data.isin([-1])

# Get columns and rows where '-1' is present
fields_with_minus_one_columns = fields_with_minus_one.any(axis=0)
fields_with_minus_one_rows = fields_with_minus_one.any(axis=1)

# Filter the data to show only rows where '-1' is present
data_with_minus_one = data[fields_with_minus_one_rows]

# Display the columns with '-1' values
columns_with_minus_one = fields_with_minus_one_columns[fields_with_minus_one_columns].index.tolist()

# Show only the relevant columns in the filtered data
data_with_minus_one_filtered = data_with_minus_one[columns_with_minus_one]


data_with_minus_one_filtered.head()
```

```python
# Ignore specific warnings
warnings.filterwarnings("ignore", category=FutureWarning, module="seaborn")
```

```python
# Encode categorical variables
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le

# Plot histograms for numerical features by gravity
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns

plt.figure(figsize=(15, 30))
for i, column in enumerate(numerical_columns, 1):
    plt.subplot(len(numerical_columns), 1, i)
    sns.histplot(data=data, x=column, hue='gravity', kde=True)
    plt.title(f'Distribution of {column} by Gravity')
plt.tight_layout()
plt.show()
```
